```python
#import the pickled object in order to segment the data set.
scaler = pickle.load(open("C:/IFOOD/scaler.pickle", 'rb'))
```

```python
pca = pickle.load(open("C:/IFOOD/pca.pickle", 'rb'))
```

```python
kmeans_pca = pickle.load(open("C:/IFOOD/kmeans_pca.pickle", 'rb'))
```

### Standardization 

```python
#standardizing using the standardization scaler
features = data[["Income", "Kidhome", "Teenhome", "Age", "marital_Married","marital_Divorced","marital_Single", "marital_Together","marital_Widow", "education_Basic", "education_Graduation","education_Master","education_PhD"  ]]
data_purchase_segm_std = scaler.transform(features)
```

### PCA 

```python
#Apply pca on the data tp obtain the 9 components for each row in the table
data_purchase_segm_pca = pca.transform(data_purchase_segm_std)
```

### K-means PCA 

```python
# Based on the principal components, we use the predict method from pca to segment the purchase data into the four segments.
data_purchase_segm_kmeans_pca = kmeans_pca.predict(data_purchase_segm_pca)
```

```python
data_purchase_predictors = data.copy()
```

```python
#Adding the segment information as a new column in the predictor dataframe
data_purchase_predictors['Segment'] = data_purchase_segm_kmeans_pca
```
